```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

sns.set(style='white',font='sans-serif',font_scale=1.5)
```

```python
from glob import glob
```

```python
names = glob('./test_alpha*.pkl')
```

```python
for name in names:
    df = pd.read_pickle(name)
    dataset = df.dataset.iloc[0]
    
    df1 = df.groupby(['alpha', 'Epoch', ]).mean()
    dfp1 = df1['val_n_triplets'].unstack().T
    dfp1 = dfp1.astype(int)
    dfp1 = dfp1[[0.0,  0.1,  1.0, 10]]

    dfp2 = df1['val_rmse'].unstack().T
    dfp2 = dfp2[[0.0, 0.1,  1.0, 10]]

    dfp1.columns = dfp1.columns.map(lambda x:'%.2f' % x)
    dfp2.columns = dfp2.columns.map(lambda x:'%.2f' % x)

    dfp1.columns.name = 'alpha'
    dfp2.columns.name = 'alpha'

    color = sns.color_palette("rainbow_r", len(dfp1.columns)) #PiYG
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18,6.5), sharex=True, sharey=False, )
    ax1, ax2 = axes        

    dfp2.rolling(20).mean().iloc[150:].plot(ax=ax1, lw = 3, color = color)

    dfp1.rolling(8).mean().plot(ax=ax2, lw = 3, color = color)

    ax1.tick_params(bottom='on', left='off',  labelleft='on', labelbottom='on', pad=-.6,)
    ax2.tick_params(bottom='on', left='off',  labelleft='on', labelbottom='on', pad=-.6,)

    #ax1.set_ylim(0.4, 1.0)

    ax1.set_ylabel('Validation RMSE')
    ax1.set_xlabel('Epochs')

    ax2.set_ylabel('Number of Valid Triplets')
    ax2.set_xlabel('Epochs')

    ax1.set_title('RMSE performance of validation set: %s' % dataset)
    ax2.set_title('Valid triplets of validation set: %s' % dataset)

    sns.despine(top=True, right=True, left=False, bottom=False)

    fig.tight_layout()
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.25)
    plt.savefig('./test_alpha_%s.png' % dataset, bbox_inches='tight', dpi=400)
```

```python
df = pd.read_pickle('./alpha_eaat3_1.0.pkl')
df1 = df.groupby(['alpha', 'Epoch', ]).mean()
dfp1 = df1['val_n_triplets'].unstack().T
dfp1 = dfp1.astype(int)
dfp1 = dfp1[[0.0, 0.01, 0.05, 0.1,  1.0]]

dfp2 = df1['val_rmse'].unstack().T
dfp2 = dfp2[[0.0, 0.01, 0.05, 0.1,  1.0]]

dfp1.columns = dfp1.columns.map(lambda x:'%.2f' % x)
dfp2.columns = dfp2.columns.map(lambda x:'%.2f' % x)

dfp1.columns.name = 'alpha'
dfp2.columns.name = 'alpha'

color = sns.color_palette("rainbow_r", len(dfp1.columns)) #PiYG
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18,6.5), sharex=True, sharey=False, )
ax1, ax2 = axes        

dfp2.rolling(20).mean().plot(ax=ax1, lw = 3, color = color)

dfp1.rolling(8).mean().plot(ax=ax2, lw = 3, color = color)

ax1.tick_params(bottom='on', left='off',  labelleft='on', labelbottom='on', pad=-.6,)
ax2.tick_params(bottom='on', left='off',  labelleft='on', labelbottom='on', pad=-.6,)

ax1.set_ylim(0.4, 1.0)

ax1.set_ylabel('Validation RMSE')
ax1.set_xlabel('Epochs')

ax2.set_ylabel('Number of Valid Triplets')
ax2.set_xlabel('Epochs')

ax1.set_title('RMSE performance of validation set')
ax2.set_title('Valid triplets of validation set')

sns.despine(top=True, right=True, left=False, bottom=False)

fig.tight_layout()
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.25)
#plt.savefig(os.path.join(save_dir, 'genotype-phenotype-10FCV_10t_valid_auc.pdf'), bbox_inches='tight', dpi=400)
```
