```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

sns.set(style='white',  font='sans-serif', font_scale=2)
```

```python
df1 = pd.read_csv('model_comparsion_115.csv', index_col=0)
df2 = pd.read_csv('model_comparsion_39.csv', index_col=0)
```

```python
df1.median()
```

```python
df2.median()
```

```python
df1['feature'] = '115'
df2['feature'] = '39'
```

```python
df
```

```python
df = df1.append(df2)
df = df[df.metric == 'test_rmse']
df = df[df.model == 'ACANet_PNA']
dfp1 = df.groupby(['feature', 'dataset']).mean().reset_index()

color = sns.color_palette("rainbow", 4)  # PiYG
sns.palplot(color)


fig, ax = plt.subplots(figsize=(16, 6))
sns.barplot(x='dataset', y='value', palette=color, 
            hue='feature', data=dfp1, ax=ax,
)


# ax.set_ylabel(label)
ax.set_xlabel('Dataset')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#ax.set_xticklabels(rotation=45, ha='right')

ax.tick_params(axis='y', left='off', labelleft='on', pad=.3,)
```
