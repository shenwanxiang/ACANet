```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import os, json
import seaborn as sns
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
from matplotlib.colors import ListedColormap, BoundaryNorm
from matplotlib import ticker

sns.set(style='white',  font='sans-serif', font_scale=2)
```

```python
data = pd.read_excel('HSSMS_datasets.xlsx', index_col=0)
data['cliff_%'] = data['% Cliff'].apply(lambda x:x.split('(')[0]).astype(int)
data['scaffold_n'] = data['Num Scaffolds'].astype(int)
data['n'] =  data['Train / Test'].apply(lambda x:x.split('(')[0]).astype(int)
```

```python
data['mol_per_scaffold'] = (data.n / data.scaffold_n)
```

```python
res = []
for dataset_name in data.index: #['CHEMBL4203_Ki']:
    
    data_dir = '../benchmark_performance/mae_aca_opt_cliff/%s' % dataset_name

    dfa = pd.read_csv(os.path.join(data_dir, 
                                       'alpha_performance.csv'), 
                          index_col = 0)
    best_alpha = dfa.iloc[dfa.rmse.idxmin()].alpha
    res.append(best_alpha)
```

```python
data['best_alpha'] = res
```

```python
thr = data['mol_per_scaffold'].median()
```

```python
data["scaffold_group"] = np.where(data["mol_per_scaffold"] <= thr, "low-scaffold", "high-scaffold")
```
