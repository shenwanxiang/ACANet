```python
from math import sqrt
import pandas as pd
import numpy as np

import torch
import torch.nn.functional as F
from rdkit import Chem
from scipy.stats import pearsonr
from tqdm import tqdm
from torch_geometric.loader import DataLoader
from torch_geometric.nn import global_mean_pool, global_max_pool

import random

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = 'white', font_scale=2)
```

```python
dfh_aca = pd.read_csv('./results/history_with_aca.csv', index_col=0)
dfh_no_aca = pd.read_csv('./results/history_no_aca.csv', index_col=0)
```

```python
dfp = dfh_aca.train_n_pos_triplets.to_frame(name = 'With ACA($\\alpha$ =1)').join(dfh_no_aca.train_n_pos_triplets.to_frame(name = 'Without ACA($\\alpha$=0)'))

fig, ax = plt.subplots(figsize=(8, 6))
dfp.plot(ax=ax)
ax.set_ylim(-100, 2100)

ax.set_xlabel('Epochs')
ax.set_ylabel('No. of mined ACTs')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

fig.tight_layout()
fig.savefig('./results/No_of_ATCs_with_without_ACA.pdf')
```

```python
dfp = dfh_aca.train_tsm.to_frame(name = 'With ACA($\\alpha$ =1)').join(dfh_no_aca.train_tsm.to_frame(name = 'Without ACA($\\alpha$=0)'))

fig, ax = plt.subplots(figsize=(8, 6))
dfp.plot(ax=ax)
ax.set_ylim(-0.5, 10)

ax.set_xlabel('Epochs')
ax.set_ylabel('TSM loss (AC-awareness)')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


fig.tight_layout()
fig.savefig('./results/TSM_loss_with_without_ACA.pdf')
```

```python
dfp = dfh_aca.train_reg.to_frame(name = 'With ACA($\\alpha$ =1)').join(dfh_no_aca.train_reg.to_frame(name = 'Without ACA($\\alpha$=0)'))

fig, ax = plt.subplots(figsize=(8, 6))
dfp.plot(ax=ax, lw=2)
ax.set_ylim(-0.0, 1)

ax.set_xlabel('Epochs')
ax.set_ylabel('MAE loss')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


fig.tight_layout()
fig.savefig('./results/MAE_loss_with_without_ACA.pdf')
```
